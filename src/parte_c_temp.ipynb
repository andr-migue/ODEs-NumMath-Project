# Parte C: Plano de Fase - Gu√≠a Te√≥rica Completa

## Sistema de Ecuaciones Diferenciales
**Modelo del autom√≥vil acoplado a un resorte con amortiguamiento:**

$$\begin{cases}
\frac{dx}{dt} = v \\
\frac{dv}{dt} = -\alpha v - \beta x
\end{cases}$$

**Donde:** $\alpha, \beta > 0$

---

## Paso 1: Encontrar los Puntos Cr√≠ticos

### 1.1 Igualar las derivadas a cero:
$$\frac{dx}{dt} = v = 0$$
$$\frac{dv}{dt} = -\alpha v - \beta x = 0$$

### 1.2 Resolver el sistema algebraico:
**De la primera ecuaci√≥n:**
$$v = 0$$

**Sustituyendo en la segunda ecuaci√≥n:**
$$-\alpha (0) - \beta x = 0$$
$$-\beta x = 0$$
$$x = 0$$

### 1.3 Punto cr√≠tico √∫nico:
$$\boxed{(x_0, v_0) = (0, 0)}$$

---

## Paso 2: Clasificar el Punto Cr√≠tico

### 2.1 Calcular la matriz Jacobiana:
**Para el sistema vectorial:** $\mathbf{F}(x,v) = \begin{pmatrix} v \\ -\alpha v - \beta x \end{pmatrix}$

**Las derivadas parciales son:**
- $\frac{\partial F_1}{\partial x} = \frac{\partial v}{\partial x} = 0$
- $\frac{\partial F_1}{\partial v} = \frac{\partial v}{\partial v} = 1$
- $\frac{\partial F_2}{\partial x} = \frac{\partial (-\alpha v - \beta x)}{\partial x} = -\beta$
- $\frac{\partial F_2}{\partial v} = \frac{\partial (-\alpha v - \beta x)}{\partial v} = -\alpha$

**Por lo tanto, la matriz Jacobiana es:**
$$\mathbf{J} = \begin{pmatrix}
\frac{\partial F_1}{\partial x} & \frac{\partial F_1}{\partial v} \\
\frac{\partial F_2}{\partial x} & \frac{\partial F_2}{\partial v}
\end{pmatrix} = \begin{pmatrix}
0 & 1 \\
-\beta & -\alpha
\end{pmatrix}$$

### 2.2 Encontrar los valores propios (eigenvalues):
**Resolver la ecuaci√≥n caracter√≠stica:** $\det(\mathbf{J} - \lambda \mathbf{I}) = 0$

$$\det\begin{pmatrix}
0 - \lambda & 1 \\
-\beta & -\alpha - \lambda
\end{pmatrix} = 0$$

$$\det\begin{pmatrix}
-\lambda & 1 \\
-\beta & -\alpha - \lambda
\end{pmatrix} = 0$$

**Calculando el determinante:**
$$(-\lambda)(-\alpha - \lambda) - (1)(-\beta) = 0$$
$$\lambda(\alpha + \lambda) + \beta = 0$$
$$\alpha\lambda + \lambda^2 + \beta = 0$$

### 2.3 Ecuaci√≥n caracter√≠stica:
$$\boxed{\lambda^2 + \alpha\lambda + \beta = 0}$$

### 2.4 Soluci√≥n usando la f√≥rmula cuadr√°tica:
$$\lambda = \frac{-\alpha \pm \sqrt{\alpha^2 - 4(1)(\beta)}}{2(1)}$$

$$\boxed{\lambda = \frac{-\alpha \pm \sqrt{\alpha^2 - 4\beta}}{2}}$$

---

## Paso 3: An√°lisis del Discriminante

**El comportamiento del sistema depende del discriminante:** $\Delta = \alpha^2 - 4\beta$

### Caso 1: $\alpha^2 - 4\beta > 0$ (Sobre-amortiguado)
- **Tipo:** Nodo estable
- **Eigenvalues:** Dos valores reales negativos distintos
$$\lambda_1 = \frac{-\alpha + \sqrt{\alpha^2 - 4\beta}}{2} < 0$$
$$\lambda_2 = \frac{-\alpha - \sqrt{\alpha^2 - 4\beta}}{2} < 0$$
- **Comportamiento:** Las trayectorias se acercan al origen sin oscilar
- **Nota:** Como $\sqrt{\alpha^2 - 4\beta} < \alpha$ (porque $\beta > 0$), ambos eigenvalues son negativos

### Caso 2: $\alpha^2 - 4\beta = 0$ (Cr√≠ticamente amortiguado)
- **Tipo:** Nodo degenerado (cr√≠tico)
- **Eigenvalue:** Un valor real negativo repetido
$$\lambda = \frac{-\alpha \pm 0}{2} = -\frac{\alpha}{2} < 0$$
- **Comportamiento:** Las trayectorias se acercan al origen r√°pidamente sin oscilar ni overshooting

### Caso 3: $\alpha^2 - 4\beta < 0$ (Sub-amortiguado)
- **Tipo:** Foco estable (espiral)
- **Eigenvalues:** Complejos conjugados con parte real negativa
$$\lambda = \frac{-\alpha \pm \sqrt{\alpha^2 - 4\beta}}{2} = \frac{-\alpha \pm i\sqrt{4\beta - \alpha^2}}{2}$$

**Escribiendo en forma est√°ndar:**
$$\lambda_{1,2} = -\frac{\alpha}{2} \pm i\frac{\sqrt{4\beta - \alpha^2}}{2}$$

**Donde:**
- **Parte real:** $\text{Re}(\lambda) = -\frac{\alpha}{2} < 0$ (estabilidad)
- **Parte imaginaria:** $\text{Im}(\lambda) = \pm\frac{\sqrt{4\beta - \alpha^2}}{2}$ (frecuencia de oscilaci√≥n)
- **Comportamiento:** Las trayectorias espiralan hacia el origen

---

## Paso 4: Construir el Plano de Fase

### 4.1 Ejes del plano de fase:
- **Eje horizontal ($x$):** Posici√≥n del autom√≥vil
- **Eje vertical ($v$):** Velocidad del autom√≥vil

### 4.2 Campo de direcciones:
**En cada punto $(x,v)$, la pendiente de la trayectoria es:**

Para $v \neq 0$:
$$\frac{dv}{dx} = \frac{\frac{dv}{dt}}{\frac{dx}{dt}} = \frac{-\alpha v - \beta x}{v}$$

**Simplificando:**
$$\frac{dv}{dx} = -\alpha - \frac{\beta x}{v}$$

### 4.3 Isoclinas especiales:
**Isoclina horizontal** ($\frac{dv}{dt} = 0$):
$$-\alpha v - \beta x = 0$$
$$\alpha v = -\beta x$$
$$\boxed{v = -\frac{\beta}{\alpha}x}$$

**Isoclina vertical** ($\frac{dx}{dt} = 0$):
$$v = 0 \quad \text{(eje horizontal)}$$

---

## Paso 5: An√°lisis de Estabilidad

### 5.1 Criterio de estabilidad:
**Como $\alpha, \beta > 0$, analizamos los eigenvalues:**

Para todos los casos:
- **Caso 1:** $\lambda_1, \lambda_2 < 0$ (reales negativos)
- **Caso 2:** $\lambda = -\frac{\alpha}{2} < 0$ (real negativo)
- **Caso 3:** $\text{Re}(\lambda) = -\frac{\alpha}{2} < 0$ (parte real negativa)

**Conclusi√≥n:** $\boxed{\text{El punto cr√≠tico } (0,0) \text{ es ASINT√ìTICAMENTE ESTABLE}}$

### 5.2 Interpretaci√≥n f√≠sica:
- **Posici√≥n de equilibrio:** $x = 0$ (resorte en posici√≥n natural)
- **Velocidad de equilibrio:** $v = 0$ (autom√≥vil en reposo)
- **Efecto del amortiguamiento:** $\alpha > 0$ disipa energ√≠a del sistema
- **Efecto del resorte:** $\beta > 0$ restaura hacia la posici√≥n de equilibrio

---

## Paso 6: Verificaci√≥n con Casos Particulares

### 6.1 Caso sin amortiguamiento ($\alpha = 0$):
**Sistema:** $\frac{dx}{dt} = v$, $\frac{dv}{dt} = -\beta x$

**Ecuaci√≥n caracter√≠stica:**
$$\lambda^2 + 0\cdot\lambda + \beta = 0$$
$$\lambda^2 + \beta = 0$$
$$\lambda = \pm i\sqrt{\beta}$$

**Resultado:** Eigenvalues puramente imaginarios ‚Üí **Centro** (√≥rbitas cerradas)
**Comportamiento:** Oscilaciones peri√≥dicas sin amortiguamiento

### 6.2 Condici√≥n de amortiguamiento cr√≠tico:
**Condici√≥n:** $\alpha^2 = 4\beta$
**Significado:** Transici√≥n entre comportamiento oscilatorio y no oscilatorio
**Ventaja:** Tiempo de respuesta √≥ptimo sin overshooting

---

## Paso 7: Construcci√≥n del Diagrama de Fase

### 7.1 Vectores propios (eigenvectors):
**Para cada eigenvalue $\lambda_i$, resolver:**
$$(\mathbf{J} - \lambda_i \mathbf{I})\mathbf{v}_i = \mathbf{0}$$

**Esto determina las direcciones principales de las trayectorias**

### 7.2 Forma de las trayectorias:
1. **Nodo estable:** Curvas que se acercan tangencialmente a las direcciones de los eigenvectors
2. **Foco estable:** Espirales logar√≠tmicas que convergen al origen
3. **Nodo cr√≠tico:** Curvas que convergen de manera √≥ptima sin overshooting

---

## Resumen Final para Transcribir

**üî¢ F√≥rmulas clave:**
1. **Sistema:** $\dot{x} = v$, $\dot{v} = -\alpha v - \beta x$
2. **Jacobiana:** $\mathbf{J} = \begin{pmatrix} 0 & 1 \\ -\beta & -\alpha \end{pmatrix}$
3. **Eigenvalues:** $\lambda = \frac{-\alpha \pm \sqrt{\alpha^2 - 4\beta}}{2}$
4. **Discriminante:** $\Delta = \alpha^2 - 4\beta$

**üìä Clasificaci√≥n:**
- $\Delta > 0$: Nodo estable (sobre-amortiguado)
- $\Delta = 0$: Nodo cr√≠tico (cr√≠ticamente amortiguado)  
- $\Delta < 0$: Foco estable (sub-amortiguado)

**‚úÖ Metodolog√≠a completa:**
1. Encontrar puntos cr√≠ticos ‚Üí $(0,0)$
2. Calcular matriz Jacobiana
3. Obtener eigenvalues
4. Clasificar seg√∫n discriminante
5. Analizar estabilidad
6. Construir plano de fase

```python
# Importaci√≥n de librer√≠as necesarias
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import matplotlib.patches as patches
from matplotlib.colors import LinearSegmentedColormap

# Configuraci√≥n para gr√°ficos de alta calidad
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12
plt.rcParams['lines.linewidth'] = 2

print("‚úÖ Librer√≠as importadas correctamente")
print("üìä Listo para el an√°lisis del plano de fase")
```

```python
# Funci√≥n para analizar el sistema y clasificar el punto cr√≠tico
def analizar_sistema(alpha, beta):
    """
    Analiza el sistema dx/dt = v, dv/dt = -Œ±v - Œ≤x
    y clasifica el punto cr√≠tico (0,0)
    """
    print(f"üîç AN√ÅLISIS DEL SISTEMA")
    print(f"Par√°metros: Œ± = {alpha}, Œ≤ = {beta}")
    print("-" * 50)
    
    # Matriz Jacobiana
    J = np.array([[0, 1], [-beta, -alpha]])
    print(f"Matriz Jacobiana J = \n{J}")
    
    # Calcular eigenvalues
    eigenvals = np.linalg.eigvals(J)
    discriminante = alpha**2 - 4*beta
    
    print(f"\nDiscriminante Œî = Œ±¬≤ - 4Œ≤ = {discriminante:.4f}")
    print(f"Eigenvalues: {eigenvals}")
    
    # Clasificaci√≥n
    if discriminante > 0:
        tipo = "NODO ESTABLE (Sobre-amortiguado)"
        lambda1 = (-alpha + np.sqrt(discriminante))/2
        lambda2 = (-alpha - np.sqrt(discriminante))/2
        print(f"Œª‚ÇÅ = {lambda1:.4f}, Œª‚ÇÇ = {lambda2:.4f}")
        print("üìà Trayectorias: Curvas que convergen sin oscilar")
    elif discriminante == 0:
        tipo = "NODO CR√çTICO (Cr√≠ticamente amortiguado)"
        lambda_rep = -alpha/2
        print(f"Œª = {lambda_rep:.4f} (repetido)")
        print("üìà Trayectorias: Convergencia √≥ptima sin overshooting")
    else:
        tipo = "FOCO ESTABLE (Sub-amortiguado)"
        parte_real = -alpha/2
        parte_imag = np.sqrt(4*beta - alpha**2)/2
        print(f"Œª = {parte_real:.4f} ¬± {parte_imag:.4f}i")
        print("üìà Trayectorias: Espirales que convergen al origen")
    
    print(f"\nüéØ CLASIFICACI√ìN: {tipo}")
    print(f"üîí ESTABILIDAD: ASINT√ìTICAMENTE ESTABLE (todas las partes reales < 0)")
    
    return J, eigenvals, tipo, discriminante

# Ejemplo de an√°lisis
print("=" * 60)
print("EJEMPLO 1: Sistema sub-amortiguado (Œ±=1, Œ≤=2)")
J1, eig1, tipo1, disc1 = analizar_sistema(alpha=1, beta=2)

print("\n" + "=" * 60)
print("EJEMPLO 2: Sistema sobre-amortiguado (Œ±=3, Œ≤=1)")
J2, eig2, tipo2, disc2 = analizar_sistema(alpha=3, beta=1)

print("\n" + "=" * 60)
print("EJEMPLO 3: Sistema cr√≠ticamente amortiguado (Œ±=2, Œ≤=1)")
J3, eig3, tipo3, disc3 = analizar_sistema(alpha=2, beta=1)
```

```python
# Funci√≥n para definir el sistema de ODEs
def sistema_ode(estado, t, alpha, beta):
    """
    Define el sistema de ODEs:
    dx/dt = v
    dv/dt = -Œ±v - Œ≤x
    """
    x, v = estado
    dxdt = v
    dvdt = -alpha * v - beta * x
    return [dxdt, dvdt]

# Funci√≥n para generar trayectorias
def generar_trayectorias(alpha, beta, condiciones_iniciales, t_max=10, num_puntos=1000):
    """
    Genera las trayectorias del sistema para diferentes condiciones iniciales
    """
    t = np.linspace(0, t_max, num_puntos)
    trayectorias = []
    
    for x0, v0 in condiciones_iniciales:
        estado_inicial = [x0, v0]
        solucion = odeint(sistema_ode, estado_inicial, t, args=(alpha, beta))
        trayectorias.append(solucion)
    
    return t, trayectorias

# Definir condiciones iniciales para diferentes casos
condiciones_iniciales = [
    (2, 0),    # Solo posici√≥n inicial
    (-2, 0),   # Posici√≥n inicial negativa
    (0, 2),    # Solo velocidad inicial
    (0, -2),   # Velocidad inicial negativa
    (1.5, 1),  # Posici√≥n y velocidad positivas
    (-1.5, -1), # Posici√≥n y velocidad negativas
    (1, -1),   # Posici√≥n positiva, velocidad negativa
    (-1, 1),   # Posici√≥n negativa, velocidad positiva
]

print("üöÄ Funciones para resoluci√≥n de ODEs implementadas")
print(f"üìã {len(condiciones_iniciales)} condiciones iniciales definidas")
```

```python
# Funci√≥n principal para crear el plano de fase
def crear_plano_fase(alpha, beta, xlim=(-3, 3), vlim=(-3, 3), density=15):
    """
    Crea una visualizaci√≥n completa del plano de fase
    """
    # Crear la figura con subplots
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle(f'Plano de Fase: dx/dt = v, dv/dt = -Œ±v - Œ≤x (Œ±={alpha}, Œ≤={beta})', 
                 fontsize=16, fontweight='bold')
    
    # SUBPLOT 1: Campo de direcciones
    x = np.linspace(xlim[0], xlim[1], density)
    v = np.linspace(vlim[0], vlim[1], density)
    X, V = np.meshgrid(x, v)
    
    # Calcular las direcciones del campo
    DX = V
    DV = -alpha * V - beta * X
    
    # Normalizar para mejor visualizaci√≥n
    M = np.sqrt(DX**2 + DV**2)
    M[M == 0] = 1  # Evitar divisi√≥n por cero
    DX_norm = DX / M
    DV_norm = DV / M
    
    ax1.quiver(X, V, DX_norm, DV_norm, M, cmap='viridis', alpha=0.7)
    ax1.set_xlabel('Posici√≥n (x)')
    ax1.set_ylabel('Velocidad (v)')
    ax1.set_title('Campo de Direcciones')
    ax1.grid(True, alpha=0.3)
    ax1.set_xlim(xlim)
    ax1.set_ylim(vlim)
    
    # Punto cr√≠tico
    ax1.plot(0, 0, 'ro', markersize=10, label='Punto cr√≠tico (0,0)')
    
    # Isoclinas
    x_iso = np.linspace(xlim[0], xlim[1], 100)
    v_iso_horizontal = -beta/alpha * x_iso  # dv/dt = 0
    
    # Solo mostrar isoclina horizontal si est√° en el rango
    if np.any((v_iso_horizontal >= vlim[0]) & (v_iso_horizontal <= vlim[1])):
        ax1.plot(x_iso, v_iso_horizontal, 'r--', linewidth=2, 
                label=f'Isoclina horizontal: v = -{beta/alpha:.2f}x')
    
    # Isoclina vertical (v = 0)
    ax1.axhline(y=0, color='b', linestyle='--', linewidth=2, 
                label='Isoclina vertical: v = 0')
    
    ax1.legend(fontsize=10)
    
    # SUBPLOT 2: Trayectorias en el plano de fase
    t, trayectorias = generar_trayectorias(alpha, beta, condiciones_iniciales)
    
    colores = plt.cm.Set3(np.linspace(0, 1, len(condiciones_iniciales)))
    
    for i, (trayectoria, color) in enumerate(zip(trayectorias, colores)):
        x_traj, v_traj = trayectoria[:, 0], trayectoria[:, 1]
        ax2.plot(x_traj, v_traj, color=color, linewidth=2, alpha=0.8)
        
        # Marcar condici√≥n inicial
        ax2.plot(x_traj[0], v_traj[0], 'o', color=color, markersize=8)
        
        # Flecha para mostrar direcci√≥n
        if len(x_traj) > 10:
            mid = len(x_traj) // 4
            dx = x_traj[mid+1] - x_traj[mid]
            dv = v_traj[mid+1] - v_traj[mid]
            ax2.arrow(x_traj[mid], v_traj[mid], dx*0.1, dv*0.1, 
                     head_width=0.1, head_length=0.1, fc=color, ec=color)
    
    ax2.plot(0, 0, 'ro', markersize=10, label='Punto cr√≠tico (0,0)')
    ax2.set_xlabel('Posici√≥n (x)')
    ax2.set_ylabel('Velocidad (v)')
    ax2.set_title('Trayectorias en el Plano de Fase')
    ax2.grid(True, alpha=0.3)
    ax2.set_xlim(xlim)
    ax2.set_ylim(vlim)
    ax2.legend()
    
    # SUBPLOT 3: Evoluci√≥n temporal de x(t)
    for i, (trayectoria, color) in enumerate(zip(trayectorias, colores)):
        x_traj = trayectoria[:, 0]
        ax3.plot(t, x_traj, color=color, linewidth=2, alpha=0.8,
                label=f'x‚ÇÄ={condiciones_iniciales[i][0]}, v‚ÇÄ={condiciones_iniciales[i][1]}')
    
    ax3.axhline(y=0, color='r', linestyle='--', alpha=0.5)
    ax3.set_xlabel('Tiempo (t)')
    ax3.set_ylabel('Posici√≥n x(t)')
    ax3.set_title('Evoluci√≥n Temporal de la Posici√≥n')
    ax3.grid(True, alpha=0.3)
    ax3.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
    
    # SUBPLOT 4: Evoluci√≥n temporal de v(t)
    for i, (trayectoria, color) in enumerate(zip(trayectorias, colores)):
        v_traj = trayectoria[:, 1]
        ax4.plot(t, v_traj, color=color, linewidth=2, alpha=0.8,
                label=f'x‚ÇÄ={condiciones_iniciales[i][0]}, v‚ÇÄ={condiciones_iniciales[i][1]}')
    
    ax4.axhline(y=0, color='r', linestyle='--', alpha=0.5)
    ax4.set_xlabel('Tiempo (t)')
    ax4.set_ylabel('Velocidad v(t)')
    ax4.set_title('Evoluci√≥n Temporal de la Velocidad')
    ax4.grid(True, alpha=0.3)
    ax4.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
    
    plt.tight_layout()
    
    # Informaci√≥n del an√°lisis
    J, eigenvals, tipo, disc = analizar_sistema(alpha, beta)
    
    # Agregar texto con informaci√≥n del an√°lisis
    info_text = f"""
    AN√ÅLISIS DEL SISTEMA:
    ‚Ä¢ Tipo: {tipo}
    ‚Ä¢ Discriminante: {disc:.3f}
    ‚Ä¢ Eigenvalues: {eigenvals}
    ‚Ä¢ Estabilidad: ASINT√ìTICAMENTE ESTABLE
    """
    
    plt.figtext(0.02, 0.02, info_text, fontsize=10, 
                bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue", alpha=0.8))
    
    return fig

print("üé® Funci√≥n de visualizaci√≥n del plano de fase creada")
print("üîß Lista para generar gr√°ficos interactivos")
```

```python
# HERRAMIENTA INTERACTIVA: Experimenta con tus propios valores
def analisis_personalizado(alpha, beta):
    """
    Funci√≥n para hacer an√°lisis personalizado con valores elegidos por el usuario
    """
    print("üî¨ AN√ÅLISIS PERSONALIZADO")
    print("=" * 50)
    
    # Realizar an√°lisis completo
    J, eigenvals, tipo, disc = analizar_sistema(alpha, beta)
    
    print(f"\nüéØ PREDICCI√ìN TE√ìRICA:")
    print(f"   Discriminante Œî = {disc:.4f}")
    if disc > 0:
        print("   ‚Üí Esperamos un NODO ESTABLE (curvas sin oscilar)")
    elif disc == 0:
        print("   ‚Üí Esperamos un NODO CR√çTICO (convergencia √≥ptima)")
    else:
        print("   ‚Üí Esperamos un FOCO ESTABLE (espirales)")
    
    print(f"\nüìä VERIFICACI√ìN GR√ÅFICA:")
    
    # Crear visualizaci√≥n
    fig = crear_plano_fase(alpha, beta)
    plt.show()
    
    return J, eigenvals, tipo, disc

# Funci√≥n auxiliar para comparar casos
def comparar_casos():
    """
    Crea una comparaci√≥n visual de los tres casos principales
    """
    fig, axes = plt.subplots(1, 3, figsize=(20, 6))
    
    casos = [
        (1, 2, "Sub-amortiguado\n(Foco Estable)"),
        (3, 1, "Sobre-amortiguado\n(Nodo Estable)"),
        (2, 1, "Cr√≠ticamente amortiguado\n(Nodo Cr√≠tico)")
    ]
    
    for i, (alpha, beta, titulo) in enumerate(casos):
        ax = axes[i]
        
        # Generar trayectorias
        t, trayectorias = generar_trayectorias(alpha, beta, condiciones_iniciales[:4])
        
        # Dibujar solo unas pocas trayectorias para claridad
        colores = ['red', 'blue', 'green', 'orange']
        for j, (trayectoria, color) in enumerate(zip(trayectorias, colores)):
            x_traj, v_traj = trayectoria[:, 0], trayectoria[:, 1]
            ax.plot(x_traj, v_traj, color=color, linewidth=2, alpha=0.8)
            ax.plot(x_traj[0], v_traj[0], 'o', color=color, markersize=8)
        
        # Punto cr√≠tico
        ax.plot(0, 0, 'ko', markersize=10)
        ax.set_xlabel('Posici√≥n (x)')
        ax.set_ylabel('Velocidad (v)') 
        ax.set_title(f'{titulo}\nŒ±={alpha}, Œ≤={beta}')
        ax.grid(True, alpha=0.3)
        ax.set_xlim(-3, 3)
        ax.set_ylim(-3, 3)
        
        # Informaci√≥n del discriminante
        disc = alpha**2 - 4*beta
        ax.text(0.02, 0.98, f'Œî = {disc:.1f}', transform=ax.transAxes, 
                verticalalignment='top', bbox=dict(boxstyle="round,pad=0.3", 
                facecolor="yellow", alpha=0.8))
    
    plt.tight_layout()
    plt.suptitle('Comparaci√≥n de los Tres Casos Principales', fontsize=16, y=1.02)
    plt.show()

# Ejemplo de uso
print("üß™ HERRAMIENTAS LISTAS PARA EXPERIMENTAR")
print("\nüìã Funciones disponibles:")
print("‚Ä¢ analisis_personalizado(alpha, beta) - An√°lisis completo con tus valores")
print("‚Ä¢ comparar_casos() - Comparaci√≥n visual de los tres casos")
print("\nüí° Ejemplos de uso:")
print("‚Ä¢ analisis_personalizado(0.5, 3)  # Sistema sub-amortiguado")
print("‚Ä¢ analisis_personalizado(4, 0.5)  # Sistema sobre-amortiguado")
print("‚Ä¢ comparar_casos()  # Comparaci√≥n de los tres casos")
```
