# Diagrama de bifurcación para el sistema $\dot{v} = rv - v^{3}$

Este sistema exhibe una **bifurcación pitchfork supercrítica** en $r = 0$.

---

## 1. Puntos de equilibrio

Partimos de:

$$
\dot v = rv - v^{3} = 0,
$$

que puede factorizarse como:

$$
v(r - v^{2}) = 0.
$$

Por tanto los equilibrios son:

$$
v_{0} = 0, \qquad v_{\pm} = \pm\sqrt{r} \quad \text{(solo si } r \ge 0 \text{)}.
$$

---

## Análisis de estabilidad mediante el método de los autovalores

Para estudiar la estabilidad de los puntos de equilibrio del sistema:

$$
\dot v = rv - v^{3},
$$

usamos el **método de los autovalores**, que consiste en linealizar el sistema alrededor de cada equilibrio y examinar el signo del autovalor.

---

### 1. Derivada del campo vectorial

La linealización del sistema alrededor de un equilibrio $v^*$ requiere la derivada:

$$
f(v) = rv - v^{3}, \qquad f'(v) = r - 3v^{2}.
$$

En un sistema unidimensional, el autovalor es simplemente:

$$
\lambda = f'(v^*).
$$

Criterio de estabilidad:

- Si $\lambda < 0$ → el equilibrio es **estable**.  
- Si $\lambda > 0$ → el equilibrio es **inestable**.  
- Si $\lambda = 0$ → se requiere análisis adicional (caso crítico).

---

### 2. Estabilidad del equilibrio $v = 0$

Evaluamos:

$$
\lambda_0 = f'(0) = r.
$$

Conclusión:

- Si $r < 0$: $\lambda_0 < 0 \Rightarrow$ **estable**.  
- Si $r > 0$: $\lambda_0 > 0 \Rightarrow$ **inestable**.  
- Si $r = 0$: $\lambda_0 = 0 \Rightarrow$ punto de **bifurcación**.

---

### 3. Estabilidad de los equilibrios $v = \pm\sqrt{r}$ (solo si $r > 0$)

Para estos valores:

$$
v^{*} = \pm\sqrt{r},
$$

evaluamos el autovalor:

$$
\lambda_{\pm} = f'(\pm\sqrt{r})
               = r - 3(\sqrt{r})^{2}
               = r - 3r
               = -2r.
$$

Como:

$$
-2r < 0 \quad \text{para todo } r > 0,
$$

tenemos que:

$$
v = \pm\sqrt{r} \quad \text{son estables para } r > 0.
$$

---

### 4. Conclusión mediante autovalores

- Para $r < 0$:  
  Solo existe $v = 0$ y es **estable**.

- Para $r = 0$:  
  El autovalor se anula → **bifurcación pitchfork supercrítica**.

- Para $r > 0$:  
  $v = 0$ es **inestable**,  
  y surgen dos nuevas ramas estables $v = \pm\sqrt{r}$.

Este análisis por autovalores confirma el comportamiento característico del *pitchfork supercrítico*:  
la estabilidad pasa de la rama trivial a dos ramas simétricas no triviales.

---

## 2. Estabilidad (resumen)

La derivada del campo respecto a $v$ es:

$$
f'(v) = \frac{\partial}{\partial v}(rv - v^{3}) = r - 3v^{2}.
$$

- Para $v = 0$: $f'(0) = r$.  
  - Si $r < 0$: $f'(0) < 0 \Rightarrow$ **estable**.  
  - Si $r > 0$: $f'(0) > 0 \Rightarrow$ **inestable**.

- Para $v = \pm\sqrt{r}$ (cuando existen):

$$
f'(\pm\sqrt{r}) = r - 3r = -2r.
$$

  - Si $r > 0$: $f'(\pm\sqrt{r}) < 0 \Rightarrow$ **estables**.

---




## 3. Diagrama de bifurcación 

```python
import numpy as np
import matplotlib.pyplot as plt

# Rango del parámetro r
r = np.linspace(-2, 2, 400)

# Ramas del equilibrio
v0 = np.zeros_like(r)                           # Rama trivial
v_plus = np.sqrt(np.clip(r, 0, None))          # Rama superior
v_minus = -np.sqrt(np.clip(r, 0, None))        # Rama inferior

plt.figure(figsize=(8, 5))

# ----- Rama v = 0 -----
# Estable para r<0 (línea sólida)
plt.plot(r[r < 0], v0[r < 0], 'b-', label="v = 0 (estable)")
# Inestable para r>0 (línea punteada)
plt.plot(r[r > 0], v0[r > 0], 'b--', label="v = 0 (inestable)")

# ----- Ramas v = ±sqrt(r) -----
# Estables para r > 0
plt.plot(r[r > 0], v_plus[r > 0], 'g-', label="v = +√r (estable)")
plt.plot(r[r > 0], v_minus[r > 0], 'g-', label="v = -√r (estable)")

# Línea vertical indicando punto de bifurcación
plt.axvline(0, color='k', linestyle=':', linewidth=1)
plt.text(0.05, 0.1, "Bifurcación\nen r = 0", transform=plt.gca().transAxes)

# Información del gráfico
plt.title("Diagrama de bifurcación para $\\dot{v} = r v - v^{3}$")
plt.xlabel("Parámetro $r$")
plt.ylabel("Equilibrio $v^{*}$")

plt.grid(True, alpha=0.3)
plt.legend(loc="upper left")

plt.show()
```
