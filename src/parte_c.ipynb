## Oscilador Amortiguado: Solución de Equilibrio
El modelo del Oscilador Amortiguado se define como un sistema de primer orden, donde el Plano de Fase está determinado por la Posición ($x$) en el eje horizontal y la Velocidad ($v$) en el eje vertical.

$$
\begin{cases}
\frac{dx}{dt} = v \\
\frac{dv}{dt} = -\alpha v - \beta x
\end{cases}
$$

Donde $\alpha > 0$ es el coeficiente de amortiguamiento (disipación de energía) y $\beta > 0$ es la constante del resorte (fuerza restauradora).

### Determinación del Punto Crítico
Para determinar los puntos de equilibrio, se exige que las tasas de cambio sean cero: $\frac{dx}{dt} = 0$ y $\frac{dv}{dt} = 0$.
1. De $\frac{dx}{dt} = v = 0$, se obtiene $v=0$.
2. Sustituyendo $v=0$ en $\frac{dv}{dt} = -\alpha v - \beta x = 0$, se obtiene $-\beta x = 0$, lo que implica $x=0$.

**Conclusión del Equilibrio:** El único punto de equilibrio es $\mathbf{(x, v) = (0, 0)}$. Esta solución de equilibrio describe la posición donde el automóvil (o masa) permanece constante, en "equilibrio" con su medio ambiente.

---

## Análisis de Estabilidad Lineal y Valores Propios
La clasificación del punto crítico $(0, 0)$ y la determinación de su estabilidad local se realiza mediante la linearización del sistema, lo cual implica calcular la **Matriz Jacobiana**.

### A. Matriz Jacobiana
Para el sistema $\mathbf{F}(x,v) = \begin{pmatrix} v \\ -\alpha v - \beta x \end{pmatrix}$, la matriz Jacobiana $\mathbf{J}$ es:
$$
\mathbf{J}(0,0) = \begin{pmatrix}
\frac{\partial F_1}{\partial x} & \frac{\partial F_1}{\partial v} \\
\frac{\partial F_2}{\partial x} & \frac{\partial F_2}{\partial v}
\end{pmatrix} = \begin{pmatrix}
0 & 1 \\
-\beta & -\alpha
\end{pmatrix}
$$

### B. Valores Propios y Estabilidad
Los valores propios $\lambda_{1,2}$ se obtienen de la ecuación característica $\det(\mathbf{J} - \lambda \mathbf{I}) = 0$, resultando en la ecuación $\lambda^2 + \alpha\lambda + \beta = 0$.
$$
\lambda_{1,2} = \frac{-\alpha \pm \sqrt{\alpha^2 - 4\beta}}{2}
$$

El **Criterio Fundamental de Estabilidad** establece que un punto de equilibrio es **asintóticamente estable** si y solo si todos los valores propios tienen parte real negativa ($\text{Re}(\lambda_i) < 0$ para todo $i$). Esto se debe a que la solución general es de la forma $\mathbf{x}(t) = \sum c_i e^{\lambda_i t} \mathbf{v}_i$. Si las partes reales son negativas, el término exponencial decae y el límite cuando $t \to \infty$ tiende a cero.

El análisis del Oscilador Amortiguado demuestra que la tasa de decaimiento está controlada por $\text{Re}(\lambda) = -\alpha/2$. Dado que $\alpha > 0$, la parte real es siempre negativa, lo que garantiza la estabilidad asintótica para todo el sistema.

---

## Clasificación Dinámica del Punto Crítico (0, 0)
La clasificación del tipo de punto crítico en el plano de fase está controlada por el **Discriminante** $\Delta = \alpha^2 - 4\beta$. Las categorías utilizadas se corresponden con la clasificación general del comportamiento de puntos críticos en 2D.

| Caso | Condición Física | Naturaleza de $\lambda$ | Clasificación Geométrica (Comportamiento) |
| :--- | :--- | :--- | :--- |
| **1. Sobre-amortiguado** | $\Delta > 0$ ($\alpha^2 > 4\beta$) | Dos reales distintos y negativos ($\lambda_1, \lambda_2 < 0$)  | **Nodo Estable (o Convergente)** |
| **2. Críticamente Amortiguado** | $\Delta = 0$ ($\alpha^2 = 4\beta$) | Real repetido y negativo ($\lambda = -\alpha/2 < 0$)  | **Nodo Crítico**  |
| **3. Sub-amortiguado** | $\Delta < 0$ ($\alpha^2 < 4\beta$) | Complejo conjugado con parte real negativa ($\text{Re}(\lambda) < 0$)  | **Foco Estable (o Espiral Convergente)** |

### A. Detalle de Clasificaciones
1. **Nodo Estable ($\Delta \ge 0$):** Este caso ocurre cuando el amortiguamiento domina, y el sistema regresa al equilibrio sin oscilaciones. Un nodo convergente es aquel donde **todas las trayectorias se aproximan al punto crítico**. En el caso $\Delta > 0$, es un nodo estable. El nodo propio o "punto estrella" y el nodo impropio son variantes geométricas de nodos.
2. **Foco Estable ($\Delta < 0$):** Este caso ocurre cuando la elasticidad domina sobre la disipación. La parte real negativa garantiza el decaimiento exponencial, mientras que la parte imaginaria determina la frecuencia de oscilación. Un **Punto Espiral Estable** (o espiral convergente) es un punto crítico asintóticamente estable alrededor del cual las trayectorias espirales se aproximan al punto.

**Conclusión de Estabilidad Global:** Debido a que los parámetros físicos $\alpha$ y $\beta$ son positivos, la parte real de los valores propios es siempre negativa, asegurando que el sistema es **globalmente asintóticamente estable**.

---

## Visualización en el Plano de Fase (Isoclinas)
El Plano de Fase $(x, v)$ permite visualizar cómo cambian los estados del sistema. En cualquier punto $(x, v)$, el **Vector de Flujo** indica la dirección y magnitud del cambio:
$$
\mathbf{F}(x,v) = \begin{pmatrix} v \\ -\alpha v - \beta x \end{pmatrix}
$$
La dirección de las trayectorias se puede analizar a través de las **Isoclinas Críticas**, líneas donde el campo direccional es puramente horizontal o vertical.

1. **Isoclina Vertical ($dx/dt = 0$):** Definida por la condición $v = 0$ (eje $x$). Físicamente, esto significa que la **posición no cambia** momentáneamente, ya que el sistema está en reposo ($v=0$).
2. **Isoclina Horizontal ($dv/dt = 0$):** Definida por la condición $-\alpha v - \beta x = 0$, o $v = -(\beta/\alpha)x$. Físicamente, esto significa que la **velocidad no cambia** momentáneamente, implicando que las fuerzas están equilibradas instantáneamente.

La intersección de estas dos isoclinas, $(0, 0)$, es el único punto donde el sistema está completamente estático, coincidiendo con el punto de equilibrio.

**Analogía Conclusiva:** El análisis de estabilidad de este sistema es como un faro que guía barcos hacia el puerto (el equilibrio). Si el faro es un **Foco Estable** (caso sub-amortiguado), los barcos se acercarán al puerto describiendo círculos cada vez más pequeños hasta anclar (oscilación amortiguada). Si el faro es un **Nodo Estable** (caso sobre-amortiguado), los barcos se acercarán al puerto de la manera más directa posible, sin oscilar, como si el agua estuviera muy espesa (amortiguamiento dominante). En ambos casos, el punto crítico es asintóticamente estable, asegurando que, independientemente del estado inicial, el sistema siempre retornará al reposo.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint


def critical_points(A: np.ndarray) -> dict:
    critical_point = (0, 0)
    eigenvalues = np.linalg.eigvals(A)
    
    l1, l2 = eigenvalues
    real_parts = [l.real for l in eigenvalues]
    has_imag = any(abs(l.imag) > 1e-10 for l in eigenvalues)
    
    if has_imag:
        if all(r < 0 for r in real_parts):
            classification = "Foco Estable"
        elif all(r > 0 for r in real_parts):
            classification = "Foco Inestable"
        else:
            classification = "Centro"
    else:
        if all(r < 0 for r in real_parts):
            classification = "Nodo Estable" if abs(l1 - l2) > 1e-10 else "Nodo Crítico"
        elif all(r > 0 for r in real_parts):
            classification = "Nodo Inestable"
        else:
            classification = "Punto Silla"
    
    is_stable = all(r < 0 for r in real_parts)
    
    return {
        'punto_critico': critical_point,
        'valores_propios': eigenvalues,
        'clasificacion': classification,
        'estable': is_stable
    }


def phase_plane(A: np.ndarray, title: str = "Plano de Fase", xlim: tuple = (-3, 3), ylim: tuple = (-3, 3)) -> None:
    fig, ax = plt.subplots(figsize=(10, 8))

    def system(state, t):
        return A @ state
    
    x = np.linspace(xlim[0], xlim[1], 15)
    y = np.linspace(ylim[0], ylim[1], 15)
    X, Y = np.meshgrid(x, y)
    
    U = A[0, 0] * X + A[0, 1] * Y
    V = A[1, 0] * X + A[1, 1] * Y
    
    M = np.sqrt(U**2 + V**2)
    M[M == 0] = 1
    U_norm, V_norm = U / M, V / M
    
    ax.quiver(X, Y, U_norm, V_norm, M, cmap='viridis', alpha=0.5, scale=35, width=0.003)
    
    t = np.linspace(0, 20, 1000)
    cmap = plt.get_cmap('tab10')
    colors = [cmap(i) for i in range(8)]
    
    for i, theta in enumerate(np.linspace(0, 2*np.pi, 8, endpoint=False)):
        x0, y0 = 2*np.cos(theta), 2*np.sin(theta)
        sol = odeint(system, [x0, y0], t)
        
        mask = (sol[:, 0] >= xlim[0]) & (sol[:, 0] <= xlim[1]) & \
               (sol[:, 1] >= ylim[0]) & (sol[:, 1] <= ylim[1])
        
        ax.plot(sol[mask, 0], sol[mask, 1], color=colors[i], linewidth=1.5, alpha=0.8)
        ax.plot(x0, y0, 'o', color=colors[i], markersize=5)
    
    info = critical_points(A)
    color = 'green' if info['estable'] else 'red'
    ax.plot(0, 0, 'o', color=color, markersize=12, markeredgecolor='black', 
            markeredgewidth=2, label=f"Equilibrio: {info['clasificacion']}")
    
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_xlabel('x', fontsize=12)
    ax.set_ylabel('v', fontsize=12)
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.grid(True, alpha=0.3)
    ax.legend(loc='upper right')
    ax.set_aspect('equal')
    
    l1, l2 = info['valores_propios']
    if abs(l1.imag) < 1e-10:
        eig_text = f"λ₁ = {l1.real:.3f}, λ₂ = {l2.real:.3f}"
    else:
        eig_text = f"λ = {l1.real:.3f} ± {abs(l1.imag):.3f}i"
    
    ax.text(0.02, 0.98, eig_text, transform=ax.transAxes, fontsize=10,
            verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))
    
    plt.tight_layout()
    plt.show()
```

```python

# Oscilador Amortiguado con diferentes parámetros
# Sistema: dx/dt = v, 
#          dv/dt = -αv - βx
# Matriz: A = [[0, 1], 
#            [-β, -α]]

# CASO 1: Sobre-amortiguado (α=4, β=1) → Δ > 0
alpha, beta = 4.0, 1.0
A1 = np.array([[0, 1], [-beta, -alpha]])

print("CASO 1: Sobre-amortiguado")
phase_plane(A1, title=f"Sobre-amortiguado (α={alpha}, β={beta})")


# CASO 2: Críticamente amortiguado (α=2, β=1) → Δ = 0
alpha, beta = 2.0, 1.0
A2 = np.array([[0, 1], [-beta, -alpha]])

print("\nCASO 2: Críticamente amortiguado")
phase_plane(A2, title=f"Críticamente amortiguado (α={alpha}, β={beta})")


# CASO 3: Sub-amortiguado (α=0.5, β=1) → Δ < 0
alpha, beta = 0.5, 1.0
A3 = np.array([[0, 1], [-beta, -alpha]])

print("\nCASO 3: Sub-amortiguado")
phase_plane(A3, title=f"Sub-amortiguado (α={alpha}, β={beta})")
```
